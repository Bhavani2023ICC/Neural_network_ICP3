a.Read the provided CSV file ‘data.csv’.
b.https://drive.google.com/drive/folders/1h8C3mLsso-R-sIOLsvoYwPLzy2fJ4IOF?usp=sharing
c.Show the basic statistical description about the data.
d.Check if the data has null values.
i.Replace the null values with the mean
e.Select at least two columns and aggregate the data using: min, max, count, mean.
f.Filter the dataframe to select the rows with calories values between 500 and 1000.
g.Filter the dataframe to select the rows with calories values > 500 and pulse < 100.
h.Create a new “df_modified” dataframe that contains all the columns from df except for “Maxpulse”.
i.Delete the “Maxpulse” column from the main df dataframe
j.Convert the datatype of Calories column to int datatype.
k.Using pandas create a scatter plot for the two columns (Duration and Calories). Example

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')

# Show the basic statistical description of the data
stats = df.describe()

# Print the statistical summary
print(stats)
# Check for null values in each column
null_values = df.isnull().sum()

# Print the number of null values in each column
print(null_values)

# Replace null values with the mean
df.fillna(df.mean(), inplace=True)

# Select at least two columns and aggregate the data using: min, max, count, mean
aggregated_data = df[['Duration', 'Calories']].agg(['min', 'max', 'count', 'mean'])
print("Aggregated data:\n", aggregated_data)

# Filter the dataframe for calories values between 500 and 1000
filtered_df_500_1000 = df[(df['Calories'] >= 500) & (df['Calories'] <= 1000)]
print("\nFiltered dataframe (Calories between 500 and 1000):\n", filtered_df_500_1000)

#  Filter the dataframe for calories > 500 and pulse < 100
filtered_df_pulse = df[(df['Calories'] > 500) & (df['Pulse'] < 100)]
print("\nFiltered dataframe (Calories > 500 and Pulse < 100):\n", filtered_df_pulse)

# Create a new “df_modified” dataframe that contains all columns except “Maxpulse”
df_modified = df.drop(columns=['Maxpulse'])
print("\nNew dataframe without Maxpulse column:\n", df_modified)

# Delete the “Maxpulse” column from the main df dataframe
df.drop(columns=['Maxpulse'], inplace=True)
print("\nOriginal dataframe after deleting Maxpulse column:\n", df)

# Convert the datatype of Calories column to int
df['Calories'] = df['Calories'].astype(int)
print("\nDataframe with Calories column as int:\n", df.dtypes)

#  Create a scatter plot for the two columns (Duration and Calories)
plt.scatter(df['Pulse'], df['Calories'])
plt.title('Scatter Plot: Pulse vs Calories')
plt.xlabel('Pulse')
plt.ylabel('Calories')
plt.grid(True)
plt.show()
```

2.Linear Regression
a)Import the given “Salary_Data.csv”
b)Split the data in train_test partitions, such that 1/3 of the data is reserved as test subset.
c)Train and predict the model.
d)Calculate the mean_squared error
e)Visualize both train and test data using scatter plot.

```python
import pandas as pd
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# a) Load the dataset
df = pd.read_csv('Salary_Data .csv')

# Assuming the dataset has two columns: 'YearsExperience' and 'Salary'
X = df[['YearsExperience']]  # Features (Independent variable)
y = df['Salary']  # Target (Dependent variable)

# b) Split the data into training (67%) and testing (33%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# c) Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the training and testing sets
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# d) Calculate the mean squared error (MSE)
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

print(f"Mean Squared Error for Training Set: {train_mse}")
print(f"Mean Squared Error for Testing Set: {test_mse}")

# e) Visualize both the train and test data using scatter plot

# Plot the training data
plt.scatter(X_train, y_train, color='blue', label='Train data')
plt.plot(X_train, y_train_pred, color='black', label='Regression line for train data')

# Plot the test data
plt.scatter(X_test, y_test, color='green', label='Test data')
plt.plot(X_test, y_test_pred, color='red', label='Regression line for test data')

# Add labels and legend
plt.title('Salary vs. Years of Experience (Train & Test)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()

```
